# 03. Advanced RAG: Re-ranking & HYDE

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.core.postprocessor import SentenceTransformerRerank

docs = SimpleDirectoryReader('../data/txt').load_data()
index = VectorStoreIndex.from_documents(docs)
reranker = SentenceTransformerRerank(model='cross-encoder/ms-marco-MiniLM-L-6-v2', top_n=5)
qe = index.as_query_engine(similarity_top_k=15, node_postprocessors=[reranker])
print(qe.query('정책 변화 핵심만 알려줘'))
```

```python
from llama_index.core.indices.query.query_transform import HyDEQueryTransform
hyde = HyDEQueryTransform()
transformed_query = hyde.run('사내 AI 보안 준수 항목은?')
print(transformed_query)
```
